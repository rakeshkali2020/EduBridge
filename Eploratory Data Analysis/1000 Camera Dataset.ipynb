# DATA VISUALIZATIONS

```python
plt.figure(figsize=(10,5))

sns.barplot(df['Price'],df['Max resolution'],color='r',label=['Price','Max resolution'])
plt.xlabel('Price')
plt.ylabel('Max resolution')
plt.title('Line Chart', loc='center')
plt.legend()
plt.xticks(rotation=70)
plt.grid(True,color='grey')
plt.show()
```

## Bar Plot showing which year has the highest camera release

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Release date', data=df, palette='coolwarm')
plt.show()
```

# Has there been an increase in Max Resolution over the years?

```python
style.use('ggplot')

plt.figure(figsize=(10,5))
x = df['Release date']
y = df['Max resolution']

plt.scatter(x, y,color='#00A36C',marker='o',linewidths=1)
plt.title('Scatter Plot of Max Resolution over the year')
plt.grid(True, color='grey')
plt.show()
```

# Has there been a decrease in Low Resolution over the years?

```python
style.use('ggplot')

plt.figure(figsize=(10,5))
x = df['Price']
y = df['Low resolution']

plt.scatter(x, y,color='#097969',marker='o',linewidths=1,s=100)
plt.title('Scatter Plot of Low Resolution over the years')
plt.grid(True,color='grey')
plt.show()
```

# Is there any relationship between Dimension and Year?

```python
style.use('ggplot')

plt.figure(figsize=(10,5))
x = df['Release date']
y = df['Dimensions']

plt.scatter(x, y,color='none',edgecolor='#001187',s=200,marker='v',linewidths=0.5)
plt.title('Scatter Plot of Low Resolution over the years')
plt.grid(True,color='#AEAEAE')
plt.show()
```

```python
storage = df['Price']
bins =[100,120,130,140,150,160,170,180,190,200]

plt.figure(figsize=(10,5))
plt.hist(storage,bins,histtype='stepfilled',color='#FF1B57',label=['Frequency','Storage Included'],alpha=1)
plt.title('Histogram Of Storage')
plt.xlabel('Storage')
plt.ylabel('Frequency')
plt.grid(True,color='#E0E0E0')
plt.legend()
plt.show()
```
