# MATPLOTLIB GRAPHS WITH ATTRIBUTES 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

# PLOTTING MUTIPLE LINE GRAPHS 

```python
style.use('ggplot')

plt.figure(figsize=(8,5))
plt.plot([1,2,3],[4,5,1],'ro--',label='line_01',linewidth=1)
plt.plot([3,2,1],[1,4,5], color='b',marker='o',label='line_02',linewidth=1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart', loc='right')
plt.legend()
plt.grid(True)
plt.show()
```

# BAR GRAPHS PLOTTING

```python
plt.figure(figsize=(8,5))

plt.bar([2,23,45,10],[20,15,22,30],label='Sample_01',alpha=0.5,width=5,linewidth=1,edgecolor = "black")
plt.bar([20,18,26,39],[29,34,56,2],label='Sample_02',alpha=0.5,width=5,linewidth=1,edgecolor = "black")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('BAR GRAPHS')
plt.legend(loc='upper right')
plt.show()
```

# SCATTER PLOT GRAPH

```python
style.use('ggplot')

color=['red','green','blue','yellow','orange','pink','magenta','violet','orangered','black']
plt.figure(figsize=(10,8))
x = np.random.randint(1,50,10)
y = np.random.randint(1,50,10)
print("The random values for: x =",x)
print("The random values for: y =", y)

plt.scatter(x, y, s=300, c=color,edgecolor='black',marker='o',linewidths=0.5, alpha=1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('SCATTER GRAPH PLOTTING')
plt.show()
```

# HISTOGRAM GRAPH PLOTTING

```python
age = np.random.randint(15,50,(50))
ml_students = np.random.randint(15,50,(50))
py_students = np.random.randint(15,50,(50))
bins =[15,20,25,30,35,40,45,50]

print(age)
print(ml_students)
print(py_students)

plt.figure(figsize=(16,8))
plt.hist([ml_students,py_students],bins,rwidth=0.6,histtype='bar',label=['ML Students','PY Students'],color=['r','b'],alpha=1)
#plt.hist(py_students,bins,rwidth=0.6,histtype='bar',label='PY Students',color='r',alpha=0.5)
plt.title('Students Involved In ML and PY Course')
plt.xlabel('Age')
plt.ylabel('No. of Students')
plt.grid(True,color='k')
plt.legend()
plt.show()
```

# PIE CHART PLOTTING

```python
labels = 'Python', 'C++', 'Ruby', 'Java','Perl'
sizes =  np.random.randint(100,500,(5))
colors = ['gold', 'yellowgreen', 'lightcoral', 'skyblue','grey']
explode = (0, 0.1, 0, 0, 0)  # explode 2nd slice

# Plot
plt.figure(figsize=(10,6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.2f%%', shadow=True, startangle=140)

plt.legend(loc="best")
plt.axis('equal')
plt.show()
```
